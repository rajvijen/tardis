## Plotting the Interactive Spectrum

### Spectrum data

````python
# Data to represent
spectrum = sim.runner.spectrum
spectrum_virtual = sim.runner.spectrum_virtual
spectrum_integrated = sim.runner.spectrum_integrated
````

### Plot

````python
# Set up interactive tools
# hover properties
props = [("Wavelength", "@{x} Å"), 
         ("Luminosity", "@{y} erg/s")
        ]
# format all props
formatters = {
    '@x': 'printf',
    '@y': 'printf'
}
# display a tooltip whenever the cursor is vertically in line with a glyph
mode='vline'

hover = HoverTool(tooltips = props, formatters = formatters, mode = mode)
TOOLS = ['pan','box_zoom','wheel_zoom',hover,'undo','redo','reset','save']

# Create the figure
p1 = figure(plot_height = 450, plot_width = 650,
    title = "Atomic data Spectrum for different Packets",
    tools = TOOLS,
    # tooltips = TOOLTIPS,
    toolbar_location='right',
    x_axis_location="above",
    x_range=(3000,9000),
    background_fill_color="#efefef")

p1.xaxis.axis_label = 'Wavelength [Å]'
p1.yaxis.axis_label = 'Luminosity [erg/s]'

p1.line(spectrum.wavelength, spectrum.luminosity_density_lambda, 
        color='blue', legend_label='normal packets')
p1.line(spectrum.wavelength, spectrum_virtual.luminosity_density_lambda, 
        color='orange', legend_label='virtual packets')
p1.line(spectrum.wavelength, spectrum_integrated.luminosity_density_lambda, 
        color='green', legend_label='formal integral')

# -------------------------------------------------------------------------------
# select for rangeTool
select = figure(title="Drag the middle and edges of the selection box to change the range above",
                plot_height=130, plot_width=800, y_range=p1.y_range,
                x_axis_type=None, y_axis_type=None,
                tools="", toolbar_location=None, background_fill_color="#efefef")

range_tool = RangeTool(x_range=p1.x_range)
range_tool.overlay.fill_color = "navy"
range_tool.overlay.fill_alpha = 0.2

select.line(spectrum.wavelength, spectrum.luminosity_density_lambda, color='blue')
select.line(spectrum.wavelength, spectrum_virtual.luminosity_density_lambda, color='orange')
select.line(spectrum.wavelength, spectrum_integrated.luminosity_density_lambda, color='green')

select.ygrid.grid_line_color = None
select.add_tools(range_tool)
select.toolbar.active_multi = range_tool

show(column(p1, select))

# Can be added to html file
# output_file("quickstart.html", title="Quickstart for TARDIS example")

# VALIDATION
# @warning(MISSING_RENDERERS)
# def _check_no_glyph_renderers(self):
#     if bad_condition: return "message"
````
